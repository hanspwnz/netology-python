**Регрессионный анализ** - можно ли провести линейную зависимость между двумя выборками. Возьмем Ladder - удовлетворение жизнью, и generosity - доброта. Мы хотим найти прямую, которая разделит точки в пространстве наполовину. Это можно сделать графически. Нарисовать график. Можно ли сказать, что данная прямая является собственным вектором? При проецировании данных на эту прямую разброс данных максимален?

```python
ladder = data['Ladder']
generosity = data['Generosity']
sns.regplot(x=ladder, y=generosity,data=data, fit_reg=True)
plt.xlabel('Удовлетворение жизнью', fontsize=16)
plt.ylabel('Доброта', fontsize=16)
plt.title('Корреляционная зависимость двух выборок', fontsize=18)
plt.grid(False)
plt.show()
```

Данная прямая являемся прямой линейной регрессии и совпадает с первым собственным вектором. Разброс данных максимален.

# Дискретное распределение случайной величины

Напишите код для моделирования 10 подбрасываний игрального кубика. Начертите диаграму histogram. Какое это распределение?

```python
ten_throws = np.random.randint(1, 7, size=10)
ten_throws
```

```python
plt.hist(ten_throws)
plt.xlabel = ('Dice value')
plt.ylabel = ('Dice value count')
plt.title = ('Dice throws hist (10 throws)')

plt.show()
```

Напишите код для моделирования 1000 подбрасываний игрального кубика. Начертите диаграму histogram. Почему график изменился? Как применить центральную предельную теорему?

```python
thousand_throws  = np.random.randint(1, 7, size=1000)
thousand_throws
```

```python
plt.hist(thousand_throws)
plt.xlabel = ('Dice value')
plt.ylabel = ('Dice value count')
plt.title = ('Dice throws hist (10 throws)')

plt.show()
```

```python
thousand_throws_grouped = []
for groups in range(100):
    temp_arr = []
    for i in range (10):
        temp_arr.append(thousand_throws[groups*10 + i])
    thousand_throws_grouped.append(sum(temp_arr) / len(temp_arr))
```

```python
plt.hist(thousand_throws_grouped)
plt.xlabel = ('Dice value')
plt.ylabel = ('Dice value count')
plt.title = ('Dice throws hist (10 throws)')

plt.show()
```

ЦПТ показывает, что среднее арифметическое значение выборок (на примере по 10) случайных величин, имитирующих бросание кубика, распределено по нормальному закону.
