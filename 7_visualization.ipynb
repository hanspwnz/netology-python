### Описание датасета

Датасет "Billboard Hot weekly charts" (https://data.world/kcmillersean/billboard-hot-100-1958-2017) представляет собой все еженедельные ТОП-100 песен с сайта **Billboard.com** за период с 08.02.1958 по 22.06.2019.

**Атрибуты таблицы**:
- **url**: ссылка на чарт Billboard.com
- **WeekID**: дата обновления чарта (каждые 7 дней)
- **Week Position**: позиция в чарте на текущую неделю
- **Song**: название песни
- **Performer**: исполнитель
- **SongID**: идентификатор песни в формате 'Song' + 'Performer', т.к. у разных исполнителей есть песни с одинаковым названием
- **Instance**: количество вхождений песни в чарты, если между попаданиями в чарты *прошло более 7 дней* (например, песня вошла в 1й чарт октября 1990 и в 3й чарт сентября 1991: Instance = 2)
- **Previous Week Position**: позиция в предыдущем чарте
- **Peak Position**: максимально взятая позиция на текущую дату
- **Weeks on Chart**: количество недель подряд в чартах

```python
# прочитаем датасет и приведем элементы WeekID к типу Datetime
songs_dataset = pd.read_csv('analyze_data/Hot Stuff.csv', parse_dates=['WeekID']) 
```

## 1. Горизонтальная столбчатая диаграмма ТОП-10 песен по общему количеству недель в чартах

Простейший вариант анализа для дальнейшей визуализации - выборка топ 10 песен по определенному критерию. Один из возможных критериев по данному набору данных - количество недель нахождения в чартах, т.к. все песни, даже занимающие топ-1 в чарте, рано или поздно выходят из чартов, скорость их выхода из чартов является показателем, насколько долго песня остается популярной.

Осуществляем выборку ТОП-10 песен и просматриваем результат:

```python
songs_dataset_grouped_by_overall_chart_duration = (
    songs_dataset.filter(items=['SongID','Song', 'Performer', 'Weeks on Chart'])
    .groupby(['SongID', 'Song', 'Performer'])
    .agg('count')
    .rename(columns={'Weeks on Chart':'Overall Weeks'})
    .reset_index()
    .sort_values(by='Overall Weeks')
)
```

```python
songs_dataset_grouped_by_overall_chart_duration.tail(10)
```

Для визуализации различия общего количества недель в чартах самых долго прослушиваемых песен строим горизонтальный барплот ТОП-10 песен как наиболее наглядно отображающий количественный признак для сравнения с остальными. По горизонтали отображается общее количество недель нахождения в чартах, по вертикали - песня и исполнитель в скобках.

```python
plt.barh(songs_dataset_grouped_by_overall_chart_duration.tail(10)['Song']
         + ' (' + songs_dataset_grouped_by_overall_chart_duration.tail(10)['Performer'] + ')',
         songs_dataset_grouped_by_overall_chart_duration.tail(10)['Overall Weeks'])
plt.grid()
plt.xlabel('Overall Weeks in charts', fontsize=16)
plt.ylabel('Songs', fontsize=16)
plt.title('Top-10 songs by overall weeks in charts', fontsize=18)
plt.show()
```

Из графика видно, что на первом месте с результатом нахождения в чартах более 80 недель находится песня "Imagine Dragons - Radioactive" с отрывом от второго места более 5 недель (8). Разница между первым и последним местом в ТОП-10 составляет более 20 недель (25) при общем среднем значении нахождения песни в чартах 9 недель.

## 2. Круговая диаграмма ТОП-10 артистов по количеству входящих в чарты песен

Одним из показателей популярности исполнителя является его прослушиваемость. Количество популярных песен может служить хорошей метрикой для оценки артиста. Для этого сгруппируем данные по исполнителю и его уникальным песням, выбросив повторяющиеся значения.

```python
songs_dataset_grouped_by_artist = (
    songs_dataset.filter(items=['Performer', 'SongID'])
    .drop_duplicates(subset='SongID', keep='first', inplace=False)
    .groupby(['Performer'])
    .agg('count')
    .rename(columns={'SongID':'Songs count'})
    .reset_index()
    .sort_values(by='Songs count')
)
```

```python
songs_dataset_grouped_by_artist.tail(10)
```

Для визуализации наиболее популярного исполнителя по сравнению с другими строим круговую диаграму ТОП-10 исполнителей, явно показывающую различия между количеством их песен, попавших в чарты. Доля в PieChrat'е - количество песен исполнителя, попавших в чарты, по отношению к общему количеству песен, попавших в чарты топ-10 исполнителей.

```python
labels = songs_dataset_grouped_by_artist.tail(10)['Performer']
sizes = songs_dataset_grouped_by_artist.tail(10)['Songs count']

plt.pie(sizes, labels=labels,
autopct = lambda p: '{:.2f}%({:.0f})'.format(p,(p/100)*sizes.sum()), shadow=True, startangle=135)

plt.axis('equal')
plt.title('Top artists by songs in charts count', fontsize=18)
plt.rcParams["figure.figsize"] = (20,10)
plt.show()
```

Как видно из графика, первое место по количеству песен в чартах занимает "группа" Glee Cast. Сериал "Glee Cast" ("Хор") был очень популярен в Америке c 2009 по 2015 годы, и множество песен из этого сериала, представляющих собой каверы известных песен, хоть раз занимало какую-либо строчку в ТОП-100 в исполнении ансамбля.

## 3. Корреляционный график зависимости количества недель нахождения песни в топе от ее максимального места

Далее проверим следующую гипотезу: чем выше максимальное место, которое занимала песня в ТОП-100, тем дольше общее время ее нахождения в чартах (в неделях).

```python
songs_dataset_grouped_by_song = (
    songs_dataset.filter(items=['Song', 'SongID', 'Peak Position'])
    .groupby(['Song', 'SongID'])
    .agg({'Peak Position':'min', 'SongID':'count'})
    .rename(columns={'SongID': 'Overall Weeks'})
    .reset_index()
)
```

```python
songs_dataset_grouped_by_song.head()
```

Для этого используем корреляционный график зависимости максимального места в чартах (по вертикали) от общего числа недель в чартах (по горизонали). Корреляционный график наглядно покажет эту зависимость и позволит проверить, что песни, находящиеся выше в топе, дольше в нем остаются, т.к. на графике отразится примерно экспоненциальное убывание значений.

```python
g = sns.lmplot(x='Overall Weeks', y='Peak Position', 
           data=songs_dataset_grouped_by_song, fit_reg=False, markers='x', height=6.27, aspect=12.7/6.27)

plt.gca().set_title('Dependency between song peak position and overall weeks in TOP', fontsize=18)
plt.gca().set_ylabel('Peak Position', fontsize=16)
plt.gca().set_xlabel('Overall Weeks', fontsize=16)
plt.show()
```

Как мы видим из корреляционного графика, ни одна из песен с пиковой позицией ниже ТОП-40 не держится в чартах дольше 35 недель, в то время, как песни с пиковой позицией ТОП-10 редко держатся в чартах меньше одной недели и достаточно часто находятся в чартах и более 40 недель.

## 4. Максимальное место в ТОП-100 по исполнителю с распределением по времени

Всю карьеру исполнителя можно представить в виде временного ряда, чтобы узнать, когда его песни, и он сам, соответственно, были на пике популярности.
Для этого строим временной ряд, где по оси x мы откладываем дату чарта, а по оси y - среднее место всех песен исполнителя, находящихся в чартах.

```python
artist_name = str(input('Введите имя исполнителя: ')) # Imagine Dragons

songs_dataset_grouped_by_artist_date = (
    songs_dataset[songs_dataset['Performer'] == artist_name]
    .filter(items=['WeekID', 'Week Position'])
    .groupby(['WeekID'])
    .agg({'Week Position':'mean'})
    .sort_values(by='WeekID')
    .reset_index()
)

g = sns.lineplot(x="WeekID", y="Week Position", data=songs_dataset_grouped_by_artist_date, marker="o")
g.set_title('Average place in TOP charts during career of ' + artist_name, fontsize=18)
g.set_ylabel('Average place', fontsize=16)
g.set_xlabel('Year', fontsize=16)

plt.gcf().set_size_inches(16.7, 8.27)
plt.gca().invert_yaxis()
```

С помощью данного графика временного ряда мы видим все пики популярности песен исполнителя (в данном случае - Madonna), а также перерывы в написании новых песен, например, с 2004 по 2006.

## 5. Распределение песен со словами "Christmas, Xmas, New Year"

Напоследок, проверим следующую гипотезу: песни со словами "Christmas, Xmas, New Year" в основном выпускают и слушают только зимой. Для этого построим барплот, на котором по оси x укажем время года, а по оси y - количество "рождественских" песен, выпущенных в это время года. Данная диаграмма позволит наглядно увидеть различие по количеству по сезонам и сделать вывод, действительно ли большинство рождественских песен популярны только зимой?

```python
def is_winter_song(x):
    if any(s in x['Song'].lower() for s in ['christmas', 'xmas', 'x-mas', 'new year']):
        if x['WeekID'].month <= 2 or x['WeekID'].month == 12:
            return 'Winter'
        elif x['WeekID'].month <= 5 or x['WeekID'].month >= 3:
            return 'Spring'
        elif x['WeekID'].month <= 8 or x['WeekID'].month >= 6:
            return 'Summer'
        else:
            return 'Autumn'
    else:
        return None
    

songs_dataset_with_xmas = songs_dataset
songs_dataset_with_xmas['song_season'] = songs_dataset_with_xmas.apply(lambda row: is_winter_song(row), axis=1)
songs_dataset_with_xmas.dropna(subset=['song_season']).reset_index()
```

```python
songs_dataset_with_xmas = (
    songs_dataset_with_xmas
    .filter(items=['song_season', 'Instance'])
    .groupby('song_season').agg('count')
    .rename(columns={'Instance':'song_season_count'})
    .reset_index()
)
```

```python
songs_dataset_with_xmas
```

```python
plt.bar(songs_dataset_with_xmas['song_season'],
         songs_dataset_with_xmas['song_season_count'])
plt.grid()
plt.xlabel('Season of year', fontsize=16)
plt.ylabel('Songs count', fontsize=16)
plt.title('Christmas songs by season', fontsize=18)
plt.show()
```

Как можно увидеть из графика, в чартах зимой находятся более 90% рождественских песен, некоторые из них попадают в весенние чарты. Из этого графика можно сделать вывод, что рождественские песни целесообразно обнародовать ближе к Новому Году или Рождеству.
